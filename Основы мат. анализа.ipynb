- **Определение функции в Python**

Соответствие $f$ между элементами множеств $X$ и $Y$ называется ***функцией***, если любому элементу $x ∈ X$ поставлен в соответствие единственный элемент $y ∈ Y$. Это обозначается следующим образом:  

$f : X → Y$ или $y = f(x), x ∈ X$.

Определения и обозначения:   
$X = D(f)$ – область определения функции $f$;  
$Y$ – область значений;  
$f(X) = E(f)$ – полный образ множества $X$ при отображении $f$;  
   
Соответствия, задаваемые функцией, удобно изображать в виде графика:  
$Γ_{f} = \left\{(x, f(x)), x ∈ X\right\}$.   
Если элементами множеств  $X$ и $Y$ являются действительные числа (т. е. $X, Y ⊂ R$), то функцию $f$ называют ***числовой функцией***.

Пусть даны две функции:   
$f : X → Y$ и $g : Y → Z.$   
Функция $h : X → Z$, определяемая равенством $h(x) = g(f(x))$, называется ***сложной функцией*** от $f$ и $g$ (или
*суперпозицией функций* $f$ и $g$, или *функцией от функции*).

- **Построение графика функции**

Графиком функции $y=f(x)$ называется множество всех точек плоскости $Oxy$, для каждой из которых $x$ является значением аргумента, а $y$ – соответствующим значением функции.   
Например, графиком функции $\sqrt{1 - x^2}$ является верхняя полуокружность радиуса $R=1$ с центром в $O(0;0)$.


![График](https://i.ibb.co/3ScsCDx/2020-06-04-18-46-13.png)

> *График функции* $\sqrt{1 - x^2}:$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-1, 1, 100)   
y = np.sqrt(1 - x**2)
plt.figure() 
plt.grid() 
plt.plot(x, y, label = '$\\sqrt{1 - x^2}$', color='b')
plt.legend(fontsize='x-large')
plt.xlabel('x')
plt.ylabel('y')
plt.show() 
```

> *График функции* $\frac{x}{x^2+10}:$

```python
x = np.arange(-5, 3, 0.1)
y = x / (x**2 + 10)
plt.figure() 
plt.grid() 
plt.plot(x, y, label = '$\\frac{x}{x^2+10}$', color='g')
plt.legend(fontsize='x-large')
plt.plot(x, y, 'g')
plt.xlabel('x')
plt.ylabel('y')
plt.show() 
```

- **Вычисление предела функции**

**Определения предела**

***На "языке последовательностей" (или по Гейне):***   
Число $A$ называется ***пределом функции $y=f(x)$ в точке*** $x_{0}$ (или при $x \rightarrow x_{0}$), если для любой последовательности допустимых значений аргумента $x_{n}$, $n \in \mathbb {N} \, (x_{n} \neq x_{0})$, сходящейся к $x_{0}$ (т.е. $\lim\limits_{x\to \infty} x_{n} = x_{0}$), последовательность соответствующих значений функций $f(x_{n})$, $n \in \mathbb {N}$, сходится к числу $A$ (т.е. $\lim\limits_{n\to \infty} f(x_{n}) = A$). 

В этом случае пишут $\lim\limits_{x \rightarrow x_{0}} f(x) = A$  или $f(x) \rightarrow A$ при $x \rightarrow x_{0}$.    
Геометрический смысл предела функции: $\lim\limits_{x \rightarrow x_{0}} f(x) = A$ означает, что для всех точек $x$, достаточно близких к точке $x_{0}$, соответствующие значения функции как угодно мало отличаются от числа $A$.


***На языке "$\varepsilon-\delta$" (или по Коши):***     
Число $A$ называется ***пределом функции в точке*** $x_{0}$ (или при $x \rightarrow x_{0}$), если для любого положительного $\varepsilon$ найдется такое положительное число $\delta$, что для всех $x \neq x_{0}$, удовлетворяющих неравенству $\left| {x - x_{0}} \right| < \delta$, выполняется неравенство $\left| f(x) - A \right| < \varepsilon$.

Записывают $\lim\limits_{x \rightarrow x_{0}} f(x) = A$.   
Это определение можно коротко записать так: $$\left( {
\forall \varepsilon  > 0 \exists \delta(\varepsilon ) > 0 \forall x: 
\underbrace{\left| {x - x_{0}} \right| < \delta, x \ne x_{0}}_{или \: 0 \: < \: \left| {x - x_{0}} \right| \: < \: \delta} 
\Longrightarrow \left| {f(x) - A}\right| < \varepsilon 
} \right) \Longleftrightarrow \lim\limits_{x \rightarrow x_{0}}f(x) = A$$

Геометрический смысл предела функции: $A = \lim\limits_{x \rightarrow x_{0}} f(x)$, если для любой $\varepsilon$-окрестности точки $A$ найдется такая $\delta$-окрестность точки $x_{0}$, что для всех $x \ne x_{0}$ из этой $\delta$-окрестности соответствующие значения функции $f(x)$ лежат в $\varepsilon$-окрестности точки $A$. Иными словами, точки графика функции $y=f(x)$ лежат внутри полосы шириной 2$\varepsilon$, ограниченной прямыми $y=A+\varepsilon, y=A-\varepsilon$. Очевидно, что величина $\delta$ зависит от выбора $\varepsilon$, поэтому пишут $\delta = \delta(\varepsilon)$.   

![График2](https://i.ibb.co/tzqzMmZ/2020-06-05-00-28-28.png)


*Пример:* Доказать, что $\lim\limits_{x \rightarrow 3} \: (2x-1) = 5$    
*Решение:* Возьмем произвольное число $\varepsilon > 0$, найдем $\delta = \delta(\varepsilon) > 0$ такое, что для всех $x$, удовлетворяющих неравенству $\left| {x - 3}\right| < \delta$, выполняется неравенство $\left| {(2x - 1)-5} \right| < \varepsilon$, т.е. $\left| {x - 3} \right| < \frac{\varepsilon}{2}$. Взяв $\delta  = \frac{\varepsilon}{2}$, видим, что для всех $x$, удовлетворяющих неравенству $\left| {x - 3}\right| < \delta \left( {=\frac{\varepsilon}{2}} \right)$ выполняется неравенство $\left| {(2x - 1)-5} \right| < \varepsilon$. Следовательно, $\lim\limits_{x \rightarrow 3} \: (2x-1) = 5$.

> *Вычисление предела $\lim\limits_{x \rightarrow 3} \: (2x-1)$ с помощью sympy.limit:*

```python
from sympy import *
x = symbols('x')
limit(2*x-1,x,3)
```

- **Нахождение производной функции (в том числе производной сложной функции)**

Пусть функция $y=f(x)$ определена на некотором интервале $(a,b)$.   
Проделаем следующие операции:  
 $-$ аргументу $x \in (a;b)$ дадим приращение $\Delta x: x + \Delta x \in (a;b)$;  
 $-$ найдем соответствующее приращение функции: $\Delta y: f(x + \Delta x) - f(x)$;     
 $-$ составим отношение приращения функции к приращению аргумента:$\frac{\Delta y}{\Delta x}$ ;      
 $-$ найдем предел этого отношения при $\Delta x \rightarrow 0: \lim\limits_{\Delta x} \frac{\Delta y}{\Delta x}$.
 
 Если этот предел существует, то его называют производной функции $f(x)$ и обозначают одним из символов $f^\prime_{x}; f^\prime(x); y^\prime; \frac{\partial y}{\partial x}; y^\prime_{x}$.    
 
 ***Производной функции*** $y=f(x)$ в точке $x_{0}$ называется предел отношения приращения функции к приращению аргумента, когда приращение аргумента стремится к нулю.

 
 Итак, по определению: $$y^\prime = \lim\limits_{\Delta x \rightarrow 0} \frac{f(x_{0} + \Delta x)-f(x_{0})}{\Delta x} \; или \; f^\prime(x_{0}) = \lim\limits_{x \rightarrow \Delta x} \frac{f(x)-f(x_{0})}{x-x_{0}}$$
  
 Производная функции $f(x)$ есть некоторая функция $f^\prime(x)$, *произведенная* из данной функции.
 
 **Пример:** Найти производную функции $y=x^2$.     
 *Решение:*    
 $-$ Аргументу $x$ даем приращение $\Delta x$;   
 $-$ находим $\Delta y: \Delta y = (x+\Delta x)^2-x^2 = 2x \cdot \Delta x+(\Delta x)^2$;    
 $-$ составляем отношение $\frac{\Delta y}{\Delta x}: \frac{\Delta y}{\Delta x} = \frac{2x \cdot \Delta x+(\Delta x)^2}{\Delta x} = 2x + \Delta x$;    
 $-$ находим предел этого отношения: $$\lim\limits_{\Delta x \rightarrow 0} \frac{\Delta y}{\Delta x} = \lim\limits_{\Delta x \rightarrow 0} (2x + \Delta x ) = 2x $$
 Таким образом, $(x^2)^\prime = 2x.$   

> *Вычисление производной $y=x^2$ с помощью sympy.diff:*

```python
y = x**2
diff(y)
```

***Сложная функция***

Пусть $y=f(u)$ и $u=\varphi(x)$, тогда $y=f(\varphi(x))$ - сложная функция  с промежуточным аргументом *u* и независимым аргументом $x$.   
Для нахождения ***сложной функции*** надо производную данной функции по промежуточному аргументу умножить на производную промежуточного аргумента по независимому аргументу.
$$y^\prime_{x} = y^\prime_{u}\cdot u^\prime_{x}$$
Это правило остается в силе, если промежуточных аргументов несколько. Так, если $y=f(u)$ и $u=\varphi(v)$ и $v=g(x)$, то $y^\prime_{x} = y^\prime_{u}\cdot u^\prime_{v}\cdot v^\prime_{x}$

*Пример:* Найдем производную сложной функции $F(x) = e^{sin(x)}$.    
*Решение:* Данная функция есть суперпозиция двух функций — экспоненты и синуса:
$F(x) = f(\varphi(x))$, где $f(y) = e^y, y = \varphi(x) = sin(x)$.    
По правилу дифференцирования сложной функции:
$F^\prime(x) = e^y(sinx)^\prime = e^{sin(x)}\cdot cos(x)$

> *Вычисление сложной производной $F(x) = e^{sin(x)}$ с помощью sympy.diff:*

```python
diff(exp(sin(x)))
```

- **Нахождение частной производной**

Пусть функция $f(x, y)$ определена в некоторой окрестности точки $(x_{0}, y_{0})$. Если существуют конечные пределы 
$$\lim\limits_{\Delta x \rightarrow 0} \: \frac{f(x_{0} + \Delta x, y_{0})-f(x_{0}, y_{0})}{\Delta x}$$
и
$$\lim\limits_{\Delta y \rightarrow 0} \: \frac{f(x_{0}, y_{0} + \Delta y)-f(x_{0}, y_{0})}{\Delta y},$$  
то их называют ***частными производными*** функции $f$ в
точке $(x_{0}, y_{0})$ по переменным $x$ и $y$ соответственно и обозначают
$\frac{\partial f}{\partial x}(x_{0}, y_{0})$ или $f^\prime_{x}(x_{0}, y_{0})$ и $\frac{\partial f}{\partial y}(x_{0}, y_{0})$ или $f^\prime_{y}(x_{0}, y_{0})$.    
Если частные производные существуют в каждой точке некоторого множества, то говорят, что функция имеет частные производные на этом множестве.      
Аналогично определяют и обозначают частные производные функции трех и более переменных. Например, если существует конечный предел
$$\lim\limits_{\Delta x_{k} \rightarrow 0} \: \frac{f(x_{1},\ldots, x_{k-1}, x_{k} + \Delta x_{k}, x_{k+1},\ldots, x_{n})}{\Delta x_{k}} - \frac{f(x_{1},\ldots, x_{k},\ldots, x_{n})}{\Delta x_{k}} $$
то его называют ***частной производной*** функции $f$ в точке
$(x_{1}, x_{2},\ldots, x_{n})$ по переменой $x_{k}$ и обозначают
$$\frac{\partial f}{\partial x_{k}}(x_{1},\ldots, x_{k},\ldots, x_{n}).$$
Для вычисления частной производной $\frac{\partial f}{\partial x_{k}}$ обычно пользуются формулами и правилами дифференцирования функции одной переменной, считая все переменные, кроме $x_{k}$, постоянными.


**Пример:** Найти частные производные первого порядка функции $z(x, y) = x^{2}-y^{2}+2xy+5$    
*Решение:*	Найдем частную производную $\frac{\partial z}{\partial x}$, при этом $y$ считаем константой.    
Получим: $$\frac{\partial z}{\partial x} = 2x+2y$$      
Аналогично находим производную $\frac{\partial z}{\partial y}$, только теперь дифференцируем по $y$ , а $x$ считаем константой:
$$\frac{\partial z}{\partial y} = 2x-2y$$   
*Ответ:* $\frac{\partial z}{\partial x} = 2x+2y;\frac{\partial z}{\partial y} = 2x-2y$



> *Поиск частных производных  функции $z(x, y) = x^{2}-y^{2}+2xy+5$:*

```python
x, y = symbols('x y')
z = x**2 - y**2 + 2*x*y + 5
print('Частная производная по x:')
diff(z, x)
```

```python
print('Частная производная по y:')
diff(z, y)
```

- **Вычисление градиента**
 
***Градиент*** — вектор, своим направлением указывающий направление наискорейшего возрастания некоторой величины $\varphi$. Другими словами, направление градиента есть направление наибыстрейшего возрастания функции. Градиент характеризует скорость изменения функции.


Градиент функции многих переменных — это вектор частных производных функции.

*для функции двух переменных формула градиента:*
$$gradf(x,y)= \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)$$
*для функции трех переменных формула градиента:*
$$gradf(x,y,z)= \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}, \frac{\partial f}{\partial z} \right)$$

Градиент часто обозначается символом «перевернутого треугольника» — оператор Набла: $∇$


Градиент дифференцируемого скалярного поля:

$$ \mathrm {grad} \,\varphi =\nabla \varphi ={\frac {\partial \varphi }{\partial x}}{\vec {i}}+{\frac {\partial \varphi }{\partial y}}{\vec {j}}+{\frac {\partial \varphi }{\partial z}}{\vec {k}}$$

Если $\varphi$  — функция $n$ переменных $x_{1},\;\ldots ,\;x_{n}$, то её градиентом называется $n$-мерный вектор

$\left({\frac {\partial \varphi }{\partial x_{1}}},\;\ldots ,\;{\frac {\partial \varphi }{\partial x_{n}}}\right),$
компоненты которого равны частным производным $\varphi$ по всем её аргументам.

Например, градиент функции $\varphi (x,\;y,\;z)=2x+3y^{2}-\sin z$ будет представлять собой:

$\nabla \varphi =\left({\frac {\partial \varphi }{\partial x}},\;{\frac {\partial \varphi }{\partial y}},\;{\frac {\partial \varphi }{\partial z}}\right)=(2,\;6y,\;-\cos z).$

> *Поиск градиента функции $\varphi (x,\;y,\;z)=2x+3y^{2}-\sin z:$*

```python
x, y, z = symbols('x y z', real=True)
f = 2*x  + 3*y**2 - sin(z)
print (diff(f, x), diff(f, y), diff(f, z), sep = ', ')
```

- **Визуализация функции нескольких переменных**

*Функция нескольких переменных*

Функцию двух переменных можно изобразить в пространстве с помощью графика.    
*Определение:* Если для каждой пары $(x,y)$ значений двух независимых переменных из некоторой области ставится в соответствие определенное значение $z$, то говорят, что $z$ является функцией двух переменных $(x,y)$ в данной области.   
*Обозначение:* $z=f(x,y)$.

Область определения (область существования) функции двух переменных $z=f(x,y)$ - это совокупность пар $(x,y)$, являющихся значениями переменных $x$ и $y$, при которых данная функция определена.

Аналогичным образом определяются функции многих переменных:  
$z=f(x_{1};x_{2};\ldots;x_{n})$

Определение 2. Пусть имеется $n$ переменных величин, и каждому набору их значений $(x_{1},x_{2},\ldots,x_{n})$ из некоторого множества $D$ соответствует одно вполне определенное значение переменной величины $z$. Тогда говорят, что задана ***функция нескольких переменных.***

> *Визуализация графика функции* $z = x^2 + y^2:$

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from pylab import meshgrid
x = np.arange(-4.0,4.0,0.01)
y = np.arange(-4.0,4.0,0.01)
X,Y = np.meshgrid(x, y)
Z = X**2 + Y**2  
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set(facecolor = 'gray')
ax.plot_surface(X, Y, Z, cmap='twilight');
```
